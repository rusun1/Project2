# Introduction

Scientific question: 
Are there similarities between cytochrome P450 enzymes that are most commonly involved in drug metabolism pathways but are from different CYP families (CYP1A2, CYP2C9, CYP2C19, CYP2D6, CYP3A4, and CYP3A5)?

The cytochrome P450 enzymes are a superfamily of monooxygenase enzymes that are essential for drug metabolism (Tuomela et al., 2011). The P450 enzymes can be inhibited or induced by drugs, and there functions and interactions with them have been widely studied. There are over 50 enzymes in this class, but 90 percent of drugs are metabolized by six specific enzymes - CYP1A2, CYP2C9, CYP2C19, CYP2D6, CYP3A4, and CYP3A5. These enzymes are found mainly in the liver, but also in the small intestine, lungs, placenta, and kidneys (Lynch et al., 2007). The US FDA has included the P450 enzymes in lableing of drugs such as diazepam, codeine, propranolo, tetrabenzanine, fluoxetine, etc. (Shankar et al., 2014). Another study found that the most common pathways of prescription drug metabolism involved CYP3A4/5, CYP2C9, CYP2D6, and CYP2C19, and CYP enzymes from families 1, 2, or 3 (McDonnell 2013). Of these enzymes, the ones that come from the same family, such as CYP3A4 and CY3A5 from the 3A family, will have similarities in structure and sequence. So the focus will be the comparison between enzymes of different families in order to find structural or sequential similarities, which might suggest why these are the enzymes most common in drug metabolism. The fasta file I created includes 13 of the cytochrome enzymes that are involved in drug metabolism pathwaysL (CYP1A1, CYP1A2, CYP1B1, CYP2A6, CYP2C8, CYP2C9, CYP3A4, CYP3A5, CYP2D6, CYP2C19, CYP2E1, CYP2B6, CYP2J2)

Hypothesis: If there are similarities in sequence and structure between CYP1A2, CYP2C9, CYP2C19, CYP2D6, CYP3A4, and CYP3A5 excluding comparison between enzymes of the same familiy, then that would suggest why these are the most common drug metabolizing enzymes.

Pairwise sequence alignment was performed between each pair of proteins to compare two sequences for their similarity. The sequences for the proteins of interested were found from Uniprot, then downloaded into a FASTA file. The alignment scores obtained from the sequence alignment was put into a matric and a heatmap was used to visualize the scores.
Homology modeling was used to compare the 3D structures of the proteins, specifically CYP1A2 and CYP3A4, and this was done in Swiss-model. Pymol was used to visualize the structural differences. 
The sequences of these enzymes can be found on uniprot (https://www.uniprot.org/) then combined into a single fasta file, and the structures can be found from RCSB (https://www.rcsb.org/) and downloaded as a pdb file.

# Performing bioinformatics analysis

#### Pairwise sequence alignment

Pairwise sequence alignment is used to identify regions of similarity between two sequences. This may indicate any functional, structural and/or evolutionary relationships between the two sequences. To perform pairwise sequence alignment, a FASTA file containing the sequences is first read in. Each sequence begins with a single-line description starting with ">" and is folllowed by the sequence data, which can be sequences of nucleotides or amino acids. The code below loads in a fasta file containing the sequences of the cytochrome P450 proteins of interest (CYP1A1, CYP1A2, CYP1B1, CYP2A6, CYP2C8, CYP2C9, CYP3A4, CYP3A5, CYP2D6, CYP2C19, CYP2E1, CYP2B6, CYP2J2). These sequences are amino acid sequences.

```python
# Read in the fasta file using SeqIO
#define the global variable Fasta
Fasta = [i for i in SeqIO.parse("cyp_p450.fasta", "fasta")]

#code check for length, the number of sequences in the fasta file
print(len(Fasta))
#code check - prints the name of the first sequence in the file
print(Fasta[0].name)

# define the global variable size
size = len(Fasta)

#define the global variable alignments_score
#initiatea an empy numpy matrix that is the length of the fasta file
alignments_score = np.empty(shape=(size,size))
```

```python
#create function to perform pairwise sequence alignment
def pairwise_seq(my_array, size):
    #Nested for loop to do pairwise sequence alignment for each pair 
    # loop over the proteins
    for i in range(size):
        # loop over the proteins again
        for j in range(size):
        # Define local variables X and Y, the two sequences to be aligned
            X = Fasta[i].seq
            Y = Fasta[j].seq
            #Code Check to see that the sequence pairs were succesfully created
            print(Fasta[i].id)
            print(Fasta[j].id)
            print(' ')
        
            #define a local variable my_array to contain alignment scores
            my_array[i,j] = int(pairwise2.align.globalxx(X, Y, score_only=True))
    
    #returns the array created
    return my_array

#define global variable My_scores to use the function defined above and create a matrix
My_scores = pairwise_seq(alignments_score, size)
print(My_scores)
```

# Plotting the results

The code below uses Matplotlib to create a heatmap. A heatmap is a data visualization technique that shows magnitude of a phenomenon as color in a two dimensional plot. Here we are using  a heatmap to compare the alignment scores obtained from the pairwise sequence alignment for each pair of sequences. The x and y axes ticks and a color bar is added, and the plot is shown. 

```python
#Create subplot
fig, ax = plt.subplots()

# Actually show the heatplot
plt.imshow(alignments_score)

# Construct the heat map to show the axes
im = ax.imshow(alignments_score)

# Add a colorbar to the right of the heatmap
cbar = ax.figure.colorbar(im)

# Show the plot
plt.show()
```

# Analysis of the results

The color bar shows the ranges of the alignment scores - sequences that have higher alignment scores are closer to yellow (>500), and sequences that have lower alignment scores are closer to dark blue (<250).

#### Homology modeling analysis

Homology modeling was performed using Swiss-model. Homology modeling is a method of visualizing the three-dimensional structure of a protein in order to compare them. Here, I chose to compare CYP1A2 and CYP3A4, two of the primary metabolizing enzymes that come from different families. The heatmap analyses above showed that the pairwise sequence alignment between CYP1A2 and CYP3A4 had a very low score of less than 250, so doinng the homology model for these would also show structural differences.

In Swiss-model, the sequence of CYP1A2 was inputed, as well as a pdb file of CYP3A4 crystal structure (1W0G) from RCSB. The figures obtained from the model results are shown below.

The 3-dimensional structure shown above shows the comparison between CYP1A2 and CYP3A4. The blue regions mean that they have high confidence scores and are similar to the original sequence. The red regions are where there are differences. The structure above has a majority of red, indicating that there are many differences between CYP1A2 and the CYP3A4 template.


The figure below shows the model-template alignment between the sequence of CYP1A2 that was inputed in the top row Model_01, and the sequence of the 1W0G.pdb as template_upload.1.A. The arrows show secondary structures and the circles shows alpha-helices. There are very few similarities in the sequences of CYP1A2 vs CYP3A4.

The figure below gives a QMEAN score, which is a quantitative score of the model. QMEAN scores that are more negative (in the blue region) are ideal, whereas more positive (red) QMEAN scores indicate lower quality. The QMEAN score obtained for this model is 0.54, which means that it is not ideal.

The figure below shows a Ramachandran plot. Ramachandran plots show the torsional angles - phi (φ)and psi (ψ) - of the residues in a peptide. It is used as a way to visualize the energetically allowed regions for the backbone dihedral angles. The top left region near 180 degrees with a large cluster of points show beta-sheets. Below thatis the right-handed alpha-helix. On the right hand side of the plot, the middle region of the y-axis shows the left-handed alpha-helix.
